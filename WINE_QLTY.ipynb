Import the required libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Reading of the file of the given dataset

```python
data=pd.read_csv("/content/wine-quality-white-and-red.csv")
data.head(24)
```

```python
set(data.quality)
```

Description of the given dataset

```python
data.describe()
```

Checking missing values in the given dataset

```python
data.isnull().sum()
```

Correlation Matrix among all features in the given dataset

```python
plt.figure(figsize=(30,20))
corr=data.corr()
sns.heatmap(corr,annot=True)
plt.savefig("Correlation Matrix.png")
plt.show()
```

Detection of outliers using z-scores

```python
from scipy import stats
num_data=data.select_dtypes(include='number')
z = np.abs(stats.zscore(num_data))
print(z)
```

z>3 Threshold -> This is an Outlier

```python
print(np.where(z>3))
```

Remove those records whose z>3 (outlier)

```python
new_data=data[(z<3).all(axis=1)]
new_data.shape
```

```python
data.shape
```

Data splitting into input features(X) and target colomn (Y)

```python
from sklearn.model_selection import train_test_split
X = new_data.drop(columns='quality')
y = new_data['quality']
X.head()
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

Data Modelling using Random forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rf_classificationModel = RandomForestClassifier(n_estimators = 100)
```

```python
X_train_encoded=pd.get_dummies(X_train)
rf_classificationModel.fit(X_train_encoded,y_train)
```

Prediction of data using Random forest Model

```python
X_test_encoded=pd.get_dummies(X_test)
y_pred=rf_classificationModel.predict(X_test_encoded)
```

```python
y_pred
```

Data Evaluation on Matrices

```python
from sklearn import metrics
print('Accuracy score', metrics.accuracy_score(y_test,y_pred))
```

Accuracy Score we got as 69.78%

Hyperparameter tuning -> Grid SearchCV

```python
plt.figure(figsize=(20,20))
from sklearn import tree
tree.plot_tree(rf_classificationModel.estimators_[0],filled=True)
```
